```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load dataset

| Feature | Data Type | Details |
| --- | --- | --- |
| Gender | String (Categorical) | Male, Female |
| Marital Status | String (Categorical) | Married, Single |
| Age | Integer (Numeric) | Age of a customer |
| Education | String (Categorical) | High School, University, Graduate, Unknown |
| Income | Integer (Numeric) | Yearly income of a customer |
| Occupation | String (Categorical) | Official, Management, Unemployed |
| Settlement Size | String (Categorical) | Small City, Mid City, Big City |

```python
df_customers = pd.read_csv('data/customers_raw.csv', header=0, index_col='CustomerID')
df_customers.head()
```

## Inspect

```python
print('The dimention of the dataset\n')
print(df_customers.shape)

print('\nDetail description of the dataset\n')
print(df_customers.info())
```

## Income Spread

```python
# plot Income spread
plt.figure(figsize=(8, 4))
sns.rugplot(x=df_customers['Income'], height=1)
plt.xlabel('Income')
plt.show()
```

```python
df_customers['Income'].describe()
```

## Income with Gender

```python
# plt.figure(figsize=(12, 8))
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
sns.histplot(data=df_customers, x='Income', kde=True, ax=ax[0])
sns.histplot(data=df_customers, x='Income', hue='Gender', kde=True, ax=ax[1])
plt.show()
```

## Age with Gender

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
sns.histplot(data=df_customers, x='Age', kde=True, ax=ax[0])
sns.histplot(data=df_customers, x='Age', kde=True, hue='Gender', ax=ax[1])
plt.show()
```

## Age vs Income with Gender

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.scatterplot(data=df_customers, x='Age', y='Income', hue='Gender', ax=ax[0])
sns.kdeplot(data=df_customers, x='Age', y='Income', hue='Gender', ax=ax[1])
plt.title('Age vs Income')
plt.show()
```

## Age vs Income with Occupation

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
sns.histplot(data=df_customers, x='Age', y='Income', hue='Occupation', ax=ax[0])
sns.kdeplot(data=df_customers, x='Age', y='Income', hue='Occupation', ax=ax[1])
# plt.savefig('output/output.png', dpi=800)
plt.show()
```

## Income of different Genders

```python
plt.figure(figsize=(8, 8))
sns.boxplot(data=df_customers, y='Income', x='Gender')
plt.xlabel('Combined')
plt.show()
```

## Income and Age of different Settlement Size

```python
# plt.figure(figsize=(8, 8))
fig, ax = plt.subplots(1, 2, figsize=(10, 6))
sns.boxplot(data=df_customers, y='Income', x='Settlement Size', ax=ax[0])
sns.boxplot(data=df_customers, y='Age', x='Settlement Size', ax=ax[1])
plt.show()
```
