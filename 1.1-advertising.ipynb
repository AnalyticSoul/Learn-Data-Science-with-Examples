```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# sns.set_theme(style="white")#, palette="pastel")
%matplotlib inline
```

```python
df_advertising = pd.read_csv('advertising.csv', index_col=0)
df_advertising.describe()
```

```python
df_advertising.isna().sum()
```

```python
plt.figure(figsize=(12, 8))
plt.xlabel('TV ad spending')
plt.ylabel('Sales')
sns.scatterplot(data=df_advertising, x='TV', y='sales')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(df_advertising[['TV']], df_advertising['sales'])
```

```python
# prediction_x = pd.DataFrame({'TV': np.linspace(0, 300, 10, dtype=int)})

predictions = lr.predict(df_advertising[['TV']])
```

```python
plt.figure(figsize=(12, 8))
plt.xlabel('TV ad spending')
plt.ylabel('Sales')
sns.scatterplot(data=df_advertising, x='TV', y='sales')
plt.plot(df_advertising['TV'], predictions, color='r')
plt.show()
```
